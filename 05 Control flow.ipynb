### 6.1 `if` Statements

````python
x = int(input('Please enter a number: '))
if x < 0:
    print('It is a negative number')
elif x == 0:
    print('Zero')
elif x == 1:
    print('Single')
else:
    print('More')
````

### 6.2 `for` Statements

````python
for i in [1, 2, 3]:
    print(i)
````

````python
for word in ['cat', 'dog', 'window', 'foo']:
    print(word, len(word))
````

````python
for i in range(5):
    print(i)
````

### 6.3 `break` and `continue` Statements

````python
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(n, 'equals', x, '*', n//x)
            break
    else:
        print(n, 'is the prime number')
````

````python
for i in ['car', 'bike', 'van', 'bus']:
    if i == 'bike':
        print('Found a bike!')
        continue
    print('Not matching... ', i)
````

````python
while(True):
    i = input("Enter 'q' to quit")
    if i == 'q':
        break
````

### 6.4 `else` in `while` and `for` loops

````python
peak = 0

while(peak < 3):
    print('Climb more!')
    peak += 1
else:
    print('Reached!')
````

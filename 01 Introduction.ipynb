### 1.1 Introduction

Python is a general purpose language. First version of Python was released in 1991. Python syntax is inspired by C syntax. If you are coming from other languages like C or Java, Please note that Python doesn't use curly braces in block of code like functions. Semicolon is also optional and generally people don't use it in Python world. Major areas that Python is being used are games, web applications, buisiness applications, scientific research, data science and testing & automation.

For more about Python history [see here](http://python-history.blogspot.com/).
